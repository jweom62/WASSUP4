# 제목과 url 데이터 목록 만들기

```python
news_title = []
news_url = []
```

```python
for item in srclist:
    title = item.text
    url = urljoin(PATH, item['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)    
```

```python
news_title
```

```python
news_url
```

## 데이터 프레임 변환

```python
import pandas as pd
```

```python
df = pd.DataFrame( {'제목':news_title, '주소':news_url}   )
df
```

```python
print(df)
```

## excel 파일 저장

```python
df.to_excel('output/naver.xlsx', index = False)
```

## 파일명에 저장날짜 추가하기

```python
# 스트링 포매팅으로 날짜 포맷을 바꿔보자.
# 2024-06-20
'%d-%02d-%02d'%(today.tm_year, today.tm_mon, today.tm_mday)
```

```python
filename = '%d-%02d-%02d'%(today.tm_year, today.tm_mon, today.tm_mday)
excel_name = filename + '.xlsx'
csv_name = filename + '.csv'
excel_name, csv_name
```
